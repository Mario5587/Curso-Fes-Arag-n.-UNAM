```python
# Ejemplo futbol

class Jugador:
    def __init__(self, nombre, equipo, posicion):
        self.nombre = nombre  # Nombre del jugador
        self.equipo = equipo  # Equipo al que pertenece
        self.posicion = posicion  # Posición en el campo
        self.goles = 0  # Número de goles

    # Método para que el jugador marque un gol
    def marcar_gol(self):
        self.goles += 1  # goles incrementan en 1

    # Método para mostrar la información del jugador
    def mostrar_info(self):
        return f"{self.nombre} juega como {self.posicion} en {self.equipo} y ha marcado {self.goles} goles"


# Creamos instancias de la clase (objetos) FUERA de la clase
messi = Jugador("Leonel Messi", "Inter Miami", "Delantero")  # Objeto 1
ronaldo = Jugador("Cristiano Ronaldo", "Al Nassr", "Delantero")  # Objeto 2
vini = Jugador("Vini Jr", "Real Madrid", "Extremo")  # Objeto 3

# Usamos métodos en los objetos
messi.marcar_gol()  # Messi Anota un gol
messi.marcar_gol()  # Messi Anota otro gol
ronaldo.marcar_gol()  # Ronaldo Anota un gol
ronaldo.marcar_gol()  # Ronaldo Anota otro gol
vini.marcar_gol()  # Vini Anota un gol

# Mostramos la información del jugador
print(messi.mostrar_info())
print(ronaldo.mostrar_info())
print(vini.mostrar_info())
```

```python
# Ejemplo de Marvel

# Clase base: Personaje (debe empezar con mayúscula)
class Personaje:
    def __init__(self, nombre, universo, habilidades):
        self.nombre = nombre
        self.universo = universo
        self.habilidades = habilidades


    def mostrar_habilidades(self):
        return f"{self.nombre} tiene las habilidades: {', '.join(self.habilidades)}"


class Superheroe(Personaje):
    def __init__(self, nombre, universo, habilidades, equipo):
        super()._init_(nombre, universo, habilidades)
        self.equipo = equipo

    def salvar_dia(self):
        return f"{self.nombre} del equipo {self.equipo} salva el día"


class Villano(Personaje):
    def __init__(self, nombre, universo, habilidades, plan_maligno):
        super().__init__(nombre, universo, habilidades)
        self.plan_maligno = plan_maligno

    def ejecutar_plan(self):
        return f"{self.nombre} está ejecutando su plan maligno: {self.plan_maligno}"


thanos = Villano("Thanos", "Marvel", ["super fuerza", "manipulación del tiempo"], "recoger las gemas del infinito")

print(thanos.ejecutar_plan())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def objective_function(x, y):
    return np.sin(np.sqrt(x*2 + y*2)) + 0.5 * np.cos(2 * x) + 0.5 * np.cos(2 * y)


num_particles = 50
iterations = 100
inertia = 0.5
cognitive = 1.5
social = 1.5


x_min, x_max = -10, 10
y_min, y_max = -10, 10
vel_min, vel_max = -1, 1


positions = np.random.uniform([x_min, y_min], [x_max, y_max], (num_particles, 2))
velocities = np.random.uniform(vel_min, vel_max, (num_particles, 2))


personal_best_positions = np.copy(positions)
personal_best_scores = objective_function(personal_best_positions[:, 0], personal_best_positions[:, 1])
global_best_position = personal_best_positions[np.argmin(personal_best_scores)]
global_best_score = np.min(personal_best_scores)


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(x_min, x_max, 100)
y = np.linspace(y_min, y_max, 100)
x, y = np.meshgrid(x, y)
z = objective_function(x, y)

ax.plot_surface(x, y, z, cmap='viridis', alpha=0.7, edgecolor='none')
ax.set_title("Particle Swarm Optimization (PSO)")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Objective Function")
particles_plot, = ax.plot([], [], [], 'bo', label="Particles", markersize=4)
ax.legend()


for iteration in range(iterations):
    scores = objective_function(positions[:, 0], positions[:, 1])


    for i in range(num_particles):
        if scores[i] < personal_best_scores[i]:
            personal_best_scores[i] = scores[i]
            personal_best_positions[i] = positions[i]

    if np.min(personal_best_scores) < global_best_score:
        global_best_score = np.min(personal_best_scores)
        global_best_position = personal_best_positions[np.argmin(personal_best_scores)]


    for i in range(num_particles):
        r1, r2 = np.random.random(2)
        velocities[i] = (inertia * velocities[i]
                         + cognitive * r1 * (personal_best_positions[i] - positions[i])
                         + social * r2 * (global_best_position - positions[i]))

        velocities[i] = np.clip(velocities[i], vel_min, vel_max)
        positions[i] += velocities[i]
        positions[i] = np.clip(positions[i], [x_min, y_min], [x_max, y_max])


    particles_plot.set_data(positions[:, 0], positions[:, 1])
    particles_plot.set_3d_properties(objective_function(positions[:, 0], positions[:, 1]))
    plt.draw()
    plt.pause(0.1)

plt.show()
```
